```python
import pandas as pd
import nltk
import matplotlib.pyplot as plt
from textblob import TextBlob

# Read CSV
df = pd.read_csv("static/data/beatles_lyrics_cleaned.csv")

# Preprocessing and Tokenization
nltk.download('punkt')
df['lyrics'] = df['lyrics'].str.lower().replace('\n', ' ').str.replace('[^\w\s]', '')
df['tokens'] = df['lyrics'].apply(nltk.word_tokenize)

# Sentiment Analysis
df['sentiment'] = df['lyrics'].apply(lambda x: TextBlob(x).sentiment.polarity)

# Frequency Analysis
word_freq = nltk.FreqDist(nltk.chain(*df['tokens']))

# You can continue with more analyses as needed
```

```python
plt.figure(figsize = (5, 5))
df = df[df.sentiment < -0.3]
plt.plot(df.sentiment, df.title, 'o')
```
